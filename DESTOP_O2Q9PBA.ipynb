```python
from pyspark.sql import SparkSession

spark= SparkSession.builder\
.appName(
    "DESTOP-O2Q9PBA")\
    .getOrCreate()

datos = [
    ("Julio", "Ventas",290), ("Luis","Administracion", 340),
    ("Marco", "Secretaria",740),("Jorge","Marketing",509),("Carlos","Comunicacion",740)
]
df=spark.createDataFrame(datos,["Nombre","Departamento","Salario"])

df.show()
df.printSchema()
df.describe().show()
```
